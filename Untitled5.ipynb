```python
# CELL 5: Explore Dataset Structure
# ========================================

def explore_dataset(df, name):
    print(f"\n=== {name} ===")
    print(f"Shape: {df.shape}")
    print(f"Columns ({len(df.columns)}): {list(df.columns)}")
    print(f"Sample data:")
    print(df.head(2))
    return df.dtypes

print("DATASET EXPLORATION:")
demo_h_types = explore_dataset(demo_h, "Demographics H (2013-2014)")
mcq_h_types = explore_dataset(mcq_h, "Medical Conditions H (2013-2014)")
rxq_h_types = explore_dataset(rxq_h, "Prescription Medications H (2013-2014)")
```

```python
#CELL 6: Identify Key Variables for Analysis
# ========================================

print("=== IDENTIFYING KEY VARIABLES ===")

# Demographics variables of interest
demo_vars = {
    'SEQN': 'Respondent sequence number',
    'RIDAGEYR': 'Age in years',
    'RIAGENDR': 'Gender',
    'RIDRETH3': 'Race/Hispanic origin w/ NH Asian',
    'DMDEDUC2': 'Education level - Adults 20+',
    'INDHHIN2': 'Annual household income',
    'DMDHRAGE': 'Age in years - HH ref person',
    'DMDHHSIZ': 'Total number of people in the Household'
}

print("Available demographic variables:")
for var, desc in demo_vars.items():
    if var in demo_h.columns:
        print(f" {var}: {desc}")
    else:
        print(f" {var}: {desc} - NOT FOUND")#
```

```python
# CELL 7: Search for Epilepsy-Related Variables
# ========================================

print("=== SEARCHING FOR EPILEPSY VARIABLES ===")

# Search for epilepsy/seizure related questions in MCQ
epilepsy_keywords = ['seizure', 'epilep', 'convul', '170', '180']

def find_epilepsy_vars(df, dataset_name):
    print(f"\nSearching in {dataset_name}:")
    found_vars = []
    for col in df.columns:
        if any(keyword.lower() in col.lower() for keyword in epilepsy_keywords):
            print(f"  Found: {col}")
            print(f"    Value counts: {df[col].value_counts()}")
            found_vars.append(col)
    return found_vars

epilepsy_vars_h = find_epilepsy_vars(mcq_h, "MCQ_H")
epilepsy_vars_i = find_epilepsy_vars(mcq_i, "MCQ_I")
```

```python
# CELL 8: Identify Common Comorbidity Variables
# ========================================

print("=== IDENTIFYING COMORBIDITY VARIABLES ===")

# Standard NHANES comorbidity questions
comorbidity_mapping = {
    'MCQ160A': 'Arthritis',
    'MCQ160B': 'Congestive Heart Failure',
    'MCQ160C': 'Coronary Heart Disease',
    'MCQ160D': 'Angina',
    'MCQ160E': 'Heart Attack',
    'MCQ160F': 'Stroke',
    'MCQ160G': 'Emphysema',
    'MCQ160K': 'Chronic Bronchitis',
    'MCQ160L': 'Liver Condition',
    'MCQ160M': 'Thyroid Problem',
    'MCQ160N': 'Kidney Disease',
    'MCQ220': 'Cancer',
    'MCQ240A': 'Diabetes'
}

def check_comorbidity_vars(df, dataset_name):
    print(f"\nComorbidity variables in {dataset_name}:")
    available_vars = {}
    for var, condition in comorbidity_mapping.items():
        if var in df.columns:
            yes_count = (df[var] == 1).sum()
            total_responses = df[var].notna().sum()
            print(f" {var} ({condition}): {yes_count}/{total_responses} positive")
            available_vars[var] = condition
        else:
            print(f" {var} ({condition}): Not found")
    return available_vars

comorbid_vars_h = check_comorbidity_vars(mcq_h, "MCQ_H")
comorbid_vars_i = check_comorbidity_vars(mcq_i, "MCQ_I")
```

```python
#CELL 9: Explore Prescription Medication Data
# ========================================

print("=== EXPLORING PRESCRIPTION MEDICATION DATA ===")

def explore_medication_data(df, dataset_name):
    print(f"\n{dataset_name} Medication Data:")
    print(f"Total medication records: {len(df)}")
    print(f"Unique patients with medications: {df['SEQN'].nunique()}")

    # Check medication name columns
    med_name_cols = [col for col in df.columns if 'DRUG' in col.upper() or 'MED' in col.upper()]
    print(f"Medication name columns: {med_name_cols}")

    if med_name_cols:
        for col in med_name_cols[:2]:  # Show first 2 columns
            print(f"\nSample {col} values:")
            print(df[col].value_counts().head())

    return med_name_cols

med_cols_h = explore_medication_data(rxq_h, "RXQ_H")
med_cols_i = explore_medication_data(rxq_i, "RXQ_I")
```

```python
#CELL 10: Search for Antiepileptic Drugs (AEDs) - FIXED VERSION
# ========================================

print("=== SEARCHING FOR ANTIEPILEPTIC DRUGS ===")

# First, let's examine the medication columns more carefully
print("Examining medication data structure:")
print(f"RXQ_H columns: {list(rxq_h.columns)}")
print(f"RXQ_I columns: {list(rxq_i.columns)}")

# Check the data types and sample values
if 'RXDDRUG' in rxq_h.columns:
    print(f"\nRXDDRUG data type: {rxq_h['RXDDRUG'].dtype}")
    print(f"Sample RXDDRUG values:")
    print(rxq_h['RXDDRUG'].head(10))
    print(f"Null values: {rxq_h['RXDDRUG'].isnull().sum()}")

# Comprehensive list of AEDs (both generic and brand names)
aed_list = [
    'phenytoin', 'dilantin', 'carbamazepine', 'tegretol', 'valproic', 'depakote',
    'lamotrigine', 'lamictal', 'levetiracetam', 'keppra', 'topiramate', 'topamax',
    'oxcarbazepine', 'trileptal', 'gabapentin', 'neurontin', 'pregabalin', 'lyrica',
    'zonisamide', 'zonegran', 'lacosamide', 'vimpat', 'eslicarbazepine', 'aptiom',
    'perampanel', 'fycompa', 'clobazam', 'onfi', 'felbamate', 'felbatol',
    'rufinamide', 'banzel', 'vigabatrin', 'sabril', 'tiagabine', 'gabitril',
    'phenobarbital', 'luminal', 'primidone', 'mysoline', 'ethosuximide', 'zarontin',
    'clonazepam', 'klonopin', 'diazepam', 'valium', 'lorazepam', 'ativan'
]

def find_aeds_safe(df, dataset_name, med_col):
    print(f"\nSearching for AEDs in {dataset_name}:")
    if med_col not in df.columns:
        print(f"Medication column {med_col} not found")
        return pd.DataFrame()

    # Check data type and handle appropriately
    print(f"Column {med_col} data type: {df[med_col].dtype}")

    # Convert to string and handle missing values
    df_copy = df.copy()
    df_copy[med_col] = df_copy[med_col].astype(str).fillna('')

    # Remove rows where medication name is empty, 'nan', or just whitespace
    valid_meds = df_copy[
        (df_copy[med_col] != '') &
        (df_copy[med_col] != 'nan') &
        (df_copy[med_col].str.strip() != '')
    ].copy()

    print(f"Total medication records: {len(df)}")
    print(f"Valid medication records: {len(valid_meds)}")

    if len(valid_meds) == 0:
        print("No valid medication names found")
        return pd.DataFrame()

    # Show sample medication names
    print(f"\nSample medication names:")
    sample_meds = valid_meds[med_col].value_counts().head(10)
    print(sample_meds)

    # Create pattern for AED search
    aed_pattern = '|'.join(aed_list)

    # Find AED medications
    aed_meds = valid_meds[valid_meds[med_col].str.contains(aed_pattern, case=False, na=False)].copy()

    print(f"\nFound {len(aed_meds)} AED prescriptions")
    print(f"Unique patients with AEDs: {aed_meds['SEQN'].nunique()}")

    if len(aed_meds) > 0:
        print(f"\nAED medications identified:")
        print(aed_meds[med_col].value_counts())

    return aed_meds

# Search for AEDs using the safe function
aeds_h = pd.DataFrame()
aeds_i = pd.DataFrame()

if 'RXDDRUG' in rxq_h.columns:
    aeds_h = find_aeds_safe(rxq_h, "RXQ_H", 'RXDDRUG')
    aeds_i = find_aeds_safe(rxq_i, "RXQ_I", 'RXDDRUG')
else:
    print("RXDDRUG column not found - checking other possible medication columns...")
    for col in rxq_h.columns:
        if 'drug' in col.lower() or 'med' in col.lower():
            print(f"Found potential medication column: {col}")
            print(f"Sample values: {rxq_h[col].head()}")

print(f"\nAED search completed.")
print(f"AEDs found in 2013-2014: {len(aeds_h)} prescriptions")
print(f"AEDs found in 2015-2016: {len(aeds_i)} prescriptions")

```

```python
# ========================================
# CELL 11: Combine Data Across Survey Cycles
# ========================================

print("=== COMBINING DATA ACROSS SURVEY CYCLES ===")

# Add cycle identifier and combine datasets
def combine_cycles(df_h, df_i, dataset_name):
    df_h_labeled = df_h.copy()
    df_h_labeled['CYCLE'] = '2013-2014'

    df_i_labeled = df_i.copy()
    df_i_labeled['CYCLE'] = '2015-2016'

    combined = pd.concat([df_h_labeled, df_i_labeled], ignore_index=True)

    print(f"{dataset_name} combined:")
    print(f"  2013-2014: {len(df_h)} records")
    print(f"  2015-2016: {len(df_i)} records")
    print(f"  Combined: {len(combined)} records")
    print(f"  Unique patients: {combined['SEQN'].nunique()}")

    return combined

# Combine all datasets
demo_combined = combine_cycles(demo_h, demo_i, "Demographics")
mcq_combined = combine_cycles(mcq_h, mcq_i, "Medical Conditions")
rxq_combined = combine_cycles(rxq_h, rxq_i, "Prescription Medications")
```

```python
 #CELL 12: Identify Epilepsy Patient Cohort
# ========================================

print("=== IDENTIFYING EPILEPSY PATIENT COHORT ===")

# Method 1: Using medical condition questionnaire (MCQ170)
epilepsy_mcq_patients = set()
if 'MCQ170' in mcq_combined.columns:
    epilepsy_mcq = mcq_combined[mcq_combined['MCQ170'] == 1]
    epilepsy_mcq_patients = set(epilepsy_mcq['SEQN'].unique())
    print(f"Patients with epilepsy from MCQ170: {len(epilepsy_mcq_patients)}")

# Method 2: Using AED prescriptions - IMPROVED VERSION
epilepsy_aed_patients = set()

# Combine AEDs from both survey cycles
all_aeds = pd.concat([aeds_h, aeds_i], ignore_index=True)

if len(all_aeds) > 0:
    epilepsy_aed_patients = set(all_aeds['SEQN'].unique())
    print(f"Patients with AED prescriptions: {len(epilepsy_aed_patients)}")

    # Show the most common AEDs found
    if 'RXDDRUG' in all_aeds.columns:
        print(f"\nMost common AEDs prescribed:")
        top_aeds = all_aeds['RXDDRUG'].value_counts().head(10)
        for med, count in top_aeds.items():
            print(f"  {med}: {count} prescriptions")
else:
    print("No AED prescriptions found - this might indicate:")
    print("1. RXDDRUG contains codes instead of drug names")
    print("2. Different column contains drug names")
    print("3. Need to use drug code lookup tables")

# Alternative: Check if there are drug code columns we should use instead
print(f"\nAlternative approach - checking for drug code columns:")
for col in rxq_combined.columns:
    if 'code' in col.lower() or 'id' in col.lower():
        print(f"Found potential code column: {col}")
        if col in rxq_combined.columns:
            print(f"  Sample values: {rxq_combined[col].value_counts().head(5)}")

# Combine both identification methods
epilepsy_all_patients = epilepsy_mcq_patients.union(epilepsy_aed_patients)
epilepsy_both_methods = epilepsy_mcq_patients.intersection(epilepsy_aed_patients)

print(f"\nEpilepsy cohort summary:")
print(f"  MCQ only: {len(epilepsy_mcq_patients - epilepsy_aed_patients)}")
print(f"  AED only: {len(epilepsy_aed_patients - epilepsy_mcq_patients)}")
print(f"  Both methods: {len(epilepsy_both_methods)}")
print(f"  Total epilepsy patients: {len(epilepsy_all_patients)}")


if len(epilepsy_aed_patients) == 0:
    print(f"\n INVESTIGATING MEDICATION DATA STRUCTURE:")
    print(f"Let's examine what the medication data actually contains...")

    # Show all columns in medication dataset
    print(f"All columns in RXQ dataset: {list(rxq_combined.columns)}")

    # Look for any text-based columns
    for col in rxq_combined.columns:
        if rxq_combined[col].dtype == 'object':
            print(f"\nText column {col} sample values:")
            print(rxq_combined[col].value_counts().head(5))
```

```python
# CELL 13: Calculate Polypharmacy Metrics
# ========================================

print("=== CALCULATING POLYPHARMACY METRICS ===")

# Count medications per patient
medication_counts = rxq_combined.groupby('SEQN').size().reset_index(name='total_medications')

print(f"Medication count statistics:")
print(medication_counts['total_medications'].describe())

# Define polypharmacy categories
def categorize_polypharmacy(med_count):
    if med_count == 0:
        return 'No medications'
    elif med_count <= 4:
        return 'Low (1-4 meds)'
    elif med_count <= 9:
        return 'Moderate (5-9 meds)'
    else:
        return 'High (10+ meds)'

medication_counts['polypharmacy_category'] = medication_counts['total_medications'].apply(categorize_polypharmacy)

print(f"\nPolypharmacy distribution:")
print(medication_counts['polypharmacy_category'].value_counts())

# Focus on epilepsy patients
epilepsy_med_counts = medication_counts[medication_counts['SEQN'].isin(epilepsy_all_patients)]
print(f"\nPolypharmacy in epilepsy patients:")
print(epilepsy_med_counts['polypharmacy_category'].value_counts())
```

```python
#CELL 14: Create Master Analysis Dataset
# ========================================

print("=== CREATING MASTER ANALYSIS DATASET ===")

# Start with demographics
master_df = demo_combined.copy()

# Add epilepsy status
master_df['has_epilepsy'] = master_df['SEQN'].isin(epilepsy_all_patients)
master_df['epilepsy_source'] = master_df['SEQN'].apply(lambda x:
    'Both' if x in epilepsy_both_methods else
    'MCQ' if x in epilepsy_mcq_patients else
    'AED' if x in epilepsy_aed_patients else
    'None')

# Add medication counts
master_df = master_df.merge(medication_counts, on='SEQN', how='left')
master_df['total_medications'] = master_df['total_medications'].fillna(0)
master_df['polypharmacy_category'] = master_df['polypharmacy_category'].fillna('No medications')

# Add comorbidities from MCQ
mcq_subset = mcq_combined[['SEQN', 'CYCLE'] + list(comorbid_vars_h.keys())].copy()
master_df = master_df.merge(mcq_subset, on=['SEQN', 'CYCLE'], how='left')

print(f"Master dataset shape: {master_df.shape}")
print(f"Epilepsy patients in master dataset: {master_df['has_epilepsy'].sum()}")
print(f"Patients with polypharmacy (≥5 meds): {(master_df['total_medications'] >= 5).sum()}")
```

```python
#CELL 17: Comprehensive Comorbidity Analysis
# ========================================

print("=== COMPREHENSIVE COMORBIDITY ANALYSIS ===")

# Create comorbidity profile for each patient
comorbidity_columns = [col for col in master_df.columns if col.startswith('MCQ') and col in comorbid_vars_h.keys()]
print(f"Available comorbidity variables: {comorbidity_columns}")

# Calculate comorbidity burden
def calculate_comorbidity_burden(row):
    """Calculate total number of comorbidities per patient"""
    return sum([1 for col in comorbidity_columns if row[col] == 1])

master_df['comorbidity_count'] = master_df[comorbidity_columns].apply(
    lambda row: sum([1 for col in comorbidity_columns if row[col] == 1]), axis=1
)

# Comorbidity categories
def categorize_comorbidity_burden(count):
    if count == 0:
        return 'No comorbidities'
    elif count <= 2:
        return 'Low (1-2)'
    elif count <= 4:
        return 'Moderate (3-4)'
    else:
        return 'High (5+)'

master_df['comorbidity_category'] = master_df['comorbidity_count'].apply(categorize_comorbidity_burden)

# Compare epilepsy vs non-epilepsy comorbidity burden
print("Comorbidity burden comparison:")
comorbidity_comparison = master_df.groupby('has_epilepsy')[['comorbidity_count', 'total_medications']].describe()
print(comorbidity_comparison)

# Statistical test for comorbidity differences
epilepsy_comorbid = master_df[master_df['has_epilepsy'] == True]['comorbidity_count']
control_comorbid = master_df[master_df['has_epilepsy'] == False]['comorbidity_count']

# Mann-Whitney U test (non-parametric)
statistic, p_value = mannwhitneyu(epilepsy_comorbid.dropna(), control_comorbid.dropna(), alternative='two-sided')
print(f"\nMann-Whitney U test for comorbidity burden:")
print(f"Statistic: {statistic}, P-value: {p_value:.6f}")
```

```python
# CELL 18: Individual Comorbidity Analysis
# ========================================

print("=== INDIVIDUAL COMORBIDITY ANALYSIS ===")

# Calculate odds ratios for each comorbidity in epilepsy patients
comorbidity_analysis = {}

for var, condition in comorbid_vars_h.items():
    if var in master_df.columns:
        # Create 2x2 contingency table
        contingency = pd.crosstab(master_df['has_epilepsy'], master_df[var] == 1)

        if contingency.shape == (2, 2):
            # Calculate odds ratio
            try:
                odds_ratio, p_val = stats.fisher_exact(contingency)

                # Calculate prevalence in each group
                epilepsy_prev = master_df[master_df['has_epilepsy'] == True][var].mean() * 100
                control_prev = master_df[master_df['has_epilepsy'] == False][var].mean() * 100

                comorbidity_analysis[condition] = {
                    'epilepsy_prevalence': epilepsy_prev,
                    'control_prevalence': control_prev,
                    'odds_ratio': odds_ratio,
                    'p_value': p_val,
                    'significant': p_val < 0.05
                }
            except:
                continue

# Create results dataframe
comorbidity_results = pd.DataFrame(comorbidity_analysis).T
comorbidity_results = comorbidity_results.sort_values('odds_ratio', ascending=False)

print("Comorbidity Analysis Results:")
print(comorbidity_results.round(3))

print(f"\nSignificant associations (p<0.05): {comorbidity_results['significant'].sum()}")
```

```python
# CELL 19: Polypharmacy Risk Analysis
# ========================================

print("=== POLYPHARMACY RISK ANALYSIS ===")

# Define high-risk polypharmacy (≥5 medications)
master_df['high_polypharmacy'] = master_df['total_medications'] >= 5

# Polypharmacy prevalence by epilepsy status
polypharmacy_crosstab = pd.crosstab(master_df['has_epilepsy'], master_df['high_polypharmacy'])
print("Polypharmacy crosstab:")
print(polypharmacy_crosstab)

# Calculate polypharmacy odds ratio
if polypharmacy_crosstab.shape == (2, 2):
    poly_or, poly_p = stats.fisher_exact(polypharmacy_crosstab)
    print(f"\nPolypharmacy in epilepsy - Odds Ratio: {poly_or:.2f}, P-value: {poly_p:.6f}")

# Analyze medication count by comorbidity burden
med_comorbid_analysis = master_df.groupby(['comorbidity_category', 'has_epilepsy'])['total_medications'].agg([
    'count', 'mean', 'std', 'median'
]).round(2)

print("\nMedication count by comorbidity burden and epilepsy status:")
print(med_comorbid_analysis)
```

```python
# CELL 31: Age and Gender Effects in Epilepsy Cohort
# ========================================

print("=== AGE AND GENDER EFFECTS IN EPILEPSY COHORT ===")

# Age group analysis
def categorize_age(age):
    if age < 18:
        return 'Pediatric (<18)'
    elif age < 65:
        return 'Adult (18-64)'
    else:
        return 'Elderly (≥65)'

epilepsy_df['age_group'] = epilepsy_df['RIDAGEYR'].apply(categorize_age)

# Age group distribution
print("Age group distribution in epilepsy patients:")
age_dist = epilepsy_df['age_group'].value_counts()
for group, count in age_dist.items():
    print(f"  {group}: {count} patients ({count/len(epilepsy_df)*100:.1f}%)")

# Comorbidity and medication patterns by age group
age_analysis = epilepsy_df.groupby('age_group').agg({
    'comorbidity_count': ['mean', 'std'],
    'total_medications': ['mean', 'std'],
    'RIDAGEYR': 'count'
}).round(2)

print(f"\nComorbidity and medication patterns by age group:")
print(age_analysis)

# Gender analysis (if available)
if 'RIAGENDR' in epilepsy_df.columns:
    gender_analysis = epilepsy_df.groupby('RIAGENDR').agg({
        'comorbidity_count': ['mean', 'std'],
        'total_medications': ['mean', 'std'],
        'RIDAGEYR': ['count', 'mean']
    }).round(2)

    print(f"\nGender differences in epilepsy cohort:")
    print(gender_analysis)
```

```python
# ========================================
# CELL 32: Machine Learning on Epilepsy Patients Only
# ========================================

print("=== MACHINE LEARNING ON EPILEPSY PATIENTS ONLY ===")

# Prepare epilepsy-only dataset for ML
epilepsy_ml_features = []

# Select relevant features for within-epilepsy analysis
feature_columns_epilepsy = ['RIDAGEYR', 'comorbidity_count'] + comorbidity_columns

# Create ML dataset for epilepsy patients only
epilepsy_ml = epilepsy_df[feature_columns_epilepsy].copy()

# Handle missing values
for col in epilepsy_ml.columns:
    if col in comorbidity_columns:
        epilepsy_ml[col] = epilepsy_ml[col].fillna(0)
    elif col in ['RIDAGEYR']:
        epilepsy_ml[col] = epilepsy_ml[col].fillna(epilepsy_ml[col].median())

# Remove rows with remaining missing values
epilepsy_ml_clean = epilepsy_ml.dropna()

print(f"Epilepsy ML dataset shape: {epilepsy_ml_clean.shape}")

# Clustering analysis for epilepsy patients only
if len(epilepsy_ml_clean) > 20:  # Need sufficient data for clustering
    from sklearn.cluster import KMeans
    from sklearn.preprocessing import StandardScaler
    from sklearn.decomposition import PCA

    # Standardize features
    epilepsy_scaler = StandardScaler()
    epilepsy_scaled = epilepsy_scaler.fit_transform(epilepsy_ml_clean)

    # Apply PCA
    epilepsy_pca = PCA(n_components=min(5, epilepsy_scaled.shape[1]))
    epilepsy_pca_data = epilepsy_pca.fit_transform(epilepsy_scaled)

    # Determine optimal clusters
    n_clusters_range = range(2, min(8, len(epilepsy_ml_clean)//5))
    if len(n_clusters_range) > 0:
        inertias = []
        for k in n_clusters_range:
            kmeans = KMeans(n_clusters=k, random_state=42)
            kmeans.fit(epilepsy_pca_data)
            inertias.append(kmeans.inertia_)

        # Use elbow method to find optimal k
        optimal_k_epilepsy = n_clusters_range[0]  # Default to minimum
        if len(inertias) > 1:
            # Simple elbow detection
            diffs = [inertias[i] - inertias[i+1] for i in range(len(inertias)-1)]
            optimal_k_epilepsy = n_clusters_range[np.argmax(diffs)]

        print(f"Optimal clusters for epilepsy patients: {optimal_k_epilepsy}")

        # Apply final clustering
        epilepsy_kmeans = KMeans(n_clusters=optimal_k_epilepsy, random_state=42)
        epilepsy_clusters = epilepsy_kmeans.fit_predict(epilepsy_pca_data)

        # Add cluster labels
        epilepsy_ml_clean['cluster'] = epilepsy_clusters

        # Analyze epilepsy clusters
        print(f"\nEpilepsy patient clusters:")
        epilepsy_cluster_analysis = epilepsy_ml_clean.groupby('cluster').agg({
            'RIDAGEYR': 'mean',
            'comorbidity_count': 'mean'
        }).round(2)

        # Add cluster sizes
        cluster_sizes = epilepsy_ml_clean['cluster'].value_counts().sort_index()
        epilepsy_cluster_analysis['patient_count'] = cluster_sizes

        print(epilepsy_cluster_analysis)

        # Characterize each cluster
        print(f"\nCluster characteristics:")
        for cluster_id in sorted(epilepsy_ml_clean['cluster'].unique()):
            cluster_data = epilepsy_ml_clean[epilepsy_ml_clean['cluster'] == cluster_id]
            avg_age = cluster_data['RIDAGEYR'].mean()
            avg_comorbid = cluster_data['comorbidity_count'].mean()
            size = len(cluster_data)

            print(f"  Cluster {cluster_id}: {size} patients, avg age {avg_age:.1f}, avg comorbidities {avg_comorbid:.1f}")

else:
    print("Insufficient data for meaningful clustering analysis")
```

```python
# CELL 33: Epilepsy-Specific Risk Stratification
# ========================================

print("=== EPILEPSY-SPECIFIC RISK STRATIFICATION ===")

# Create risk categories based on multiple factors
def calculate_risk_score(row):
    """Calculate composite risk score for epilepsy patients"""
    score = 0

    # Age risk (higher risk for elderly)
    if row['RIDAGEYR'] >= 65:
        score += 2
    elif row['RIDAGEYR'] >= 50:
        score += 1

    # Comorbidity burden
    if row['comorbidity_count'] >= 4:
        score += 3
    elif row['comorbidity_count'] >= 2:
        score += 2
    elif row['comorbidity_count'] >= 1:
        score += 1

    # Polypharmacy risk
    if row['total_medications'] >= 10:
        score += 3
    elif row['total_medications'] >= 5:
        score += 2
    elif row['total_medications'] >= 3:
        score += 1

    return score

epilepsy_df['risk_score'] = epilepsy_df.apply(calculate_risk_score, axis=1)

# Risk categories
def categorize_risk(score):
    if score <= 2:
        return 'Low Risk'
    elif score <= 4:
        return 'Moderate Risk'
    else:
        return 'High Risk'

epilepsy_df['risk_category'] = epilepsy_df['risk_score'].apply(categorize_risk)

# Risk distribution
print("Risk stratification in epilepsy patients:")
risk_dist = epilepsy_df['risk_category'].value_counts()
for category, count in risk_dist.items():
    print(f"  {category}: {count} patients ({count/len(epilepsy_df)*100:.1f}%)")

# Characteristics by risk category
risk_characteristics = epilepsy_df.groupby('risk_category').agg({
    'RIDAGEYR': ['mean', 'std'],
    'comorbidity_count': ['mean', 'std'],
    'total_medications': ['mean', 'std'],
    'risk_score': 'mean'
}).round(2)

print(f"\nCharacteristics by risk category:")
print(risk_characteristics)
```

```python
# ========================================
# CELL 41: Clinical Decision Support Visualizations
# ========================================

print("=== CLINICAL DECISION SUPPORT VISUALIZATIONS ===")

# 1. Risk matrix heatmap
fig, ax = plt.subplots(figsize=(12, 8))

# Create risk matrix based on comorbidities and medications
comorbid_bins = pd.cut(epilepsy_df['comorbidity_count'], bins=[0, 1, 3, 10],
                       labels=['0-1', '2-3', '4+'])
med_bins = pd.cut(epilepsy_df['total_medications'], bins=[0, 2, 5, 20],
                  labels=['0-2', '3-5', '6+'])

risk_matrix = pd.crosstab(comorbid_bins, med_bins, normalize='all') * 100

sns.heatmap(risk_matrix, annot=True, fmt='.1f', cmap='Reds',
            cbar_kws={'label': 'Percentage of Patients'})
ax.set_xlabel('Number of Medications')
ax.set_ylabel('Number of Comorbidities')
ax.set_title('Patient Distribution Matrix: Comorbidities vs Medications',
             fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# 2. Treatment complexity timeline (by age)
plt.figure(figsize=(14, 8))

# Group by age deciles for smoother visualization
epilepsy_df['age_decile'] = pd.cut(epilepsy_df['RIDAGEYR'], bins=10)
age_complexity = epilepsy_df.groupby('age_decile').agg({
    'comorbidity_count': 'mean',
    'total_medications': 'mean'
}).reset_index()

age_complexity['age_midpoint'] = age_complexity['age_decile'].apply(
    lambda x: x.mid if pd.notna(x) else np.nan)

plt.plot(age_complexity['age_midpoint'], age_complexity['comorbidity_count'],
         'o-', label='Mean Comorbidities', linewidth=2, markersize=8)
plt.plot(age_complexity['age_midpoint'], age_complexity['total_medications'],
         's-', label='Mean Medications', linewidth=2, markersize=8)

plt.xlabel('Age (years)')
plt.ylabel('Count')
plt.title('Treatment Complexity by Age in Epilepsy Patients', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# 3. Predictive risk visualization
plt.figure(figsize=(12, 8))

# Create risk prediction based on multiple factors
risk_factors = epilepsy_df[['RIDAGEYR', 'comorbidity_count', 'total_medications', 'risk_score']]

# Normalize factors for radar chart
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
risk_factors_norm = pd.DataFrame(scaler.fit_transform(risk_factors),
                                columns=risk_factors.columns)
risk_factors_norm['risk_category'] = epilepsy_df['risk_category']

# Plot mean values by risk category
fig, axes = plt.subplots(1, 3, figsize=(18, 6), subplot_kw=dict(projection='polar'))

risk_cats = ['Low Risk', 'Moderate Risk', 'High Risk']
colors = ['green', 'orange', 'red']

for i, (cat, color) in enumerate(zip(risk_cats, colors)):
    if cat in risk_factors_norm['risk_category'].values:
        cat_data = risk_factors_norm[risk_factors_norm['risk_category'] == cat]
        mean_values = cat_data[['RIDAGEYR', 'comorbidity_count', 'total_medications', 'risk_score']].mean()

        # Radar chart
        angles = np.linspace(0, 2*np.pi, len(mean_values), endpoint=False).tolist()
        values = mean_values.tolist()

        angles += angles[:1]  # Complete the circle
        values += values[:1]

        axes[i].plot(angles, values, 'o-', linewidth=2, color=color)
        axes[i].fill(angles, values, alpha=0.25, color=color)
        axes[i].set_xticks(angles[:-1])
        axes[i].set_xticklabels(['Age', 'Comorbidities', 'Medications', 'Risk Score'])
        axes[i].set_title(f'{cat} Profile', color=color, fontweight='bold')
        axes[i].set_ylim(0, 1)

plt.tight_layout()
plt.show()

```
